Quando trabalhamos com mapas interativos √© interessante mostrarmos o contorno de uma regi√£o, como por exemplo o contorno de um pa√≠s, Estado ou munic√≠pio. Assim, conseguimos dar enf√¢se na regi√£o de interesse.

O GEE tem uma base de dados dos contornos dos pa√≠ses para o mundo todo. Esta base de dados √© uma Feature Collection de [n√≠vel 0](https://developers.google.com/earth-engine/datasets/catalog/FAO_GAUL_2015_level0). Precisaremos informar na FeatureCollection o **Earth Engine Snippet**, como `"FAO/GAUL/2015/level0"`.

Em seguida, para selecionar um pa√≠s em espec√≠fico utilizaremos o comando `Filter` e iremos inserir o nome do pa√≠s (sem acentos e em ingl√™s). Neste exemplo queremos analisar o Brasil, assim iremos inserir o nome `Brazil`.

```python
# carregando os contornos de todos os pa√≠ses do mundo
paises = ee.FeatureCollection('FAO/GAUL/2015/level0')
```

```python
# mostrando as informa√ß√µes dos 5 primeiros pa√≠ses da featurecollection
paises.limit(5)
```

```python
# selecionando o contorno do Brasil
brasil = paises.filter(ee.Filter.eq('ADM0_NAME', 'Brazil'))
```

```python
# mostrando as propriedades do contorno do Brasil
brasil
```

Iremos adicionar o contorno do Brasil atrav√©s da fun√ß√£o do geemap chamada `Map.addLayer()`. Como argumento inseriremos a vari√°vel **brazil** e definiremos o estilo do contorno e o nome da legenda.

Al√©m disso, nesse mapa iremos usar a fun√ß√£o `centerObject()` para centrar o mapa no Brasil e aplicaremos um zoom de valor 4.

```python
# criando mapa interativo
Map = geemap.Map()

# centra o mapa na regi√£o
Map.centerObject(brasil, zoom=4)

# adicionando basemap
Map.add_basemap('Esri.WorldImagery')

# contorno de itajuba
style = {'color': 'yellow', 'fillColor': '00000000'}
Map.addLayer(brasil.style(**style), {}, 'Brasil')

# exibe na tela
Map
```

Vamos aplicar uma metodologia semelhante ao utilizado anteriormente, s√≥ que agora para escolher o contorno por Estado. No caso, iremos escolher o Estado do `Rio Grande do Sul (RS)`. Os contornos dos Estados est√£o armazenados nos dados de level 1.

O GEE tamb√©m possui uma base de dados dos Estados. Esta base de dados √© uma Feature Collection de [n√≠vel 1](https://developers.google.com/earth-engine/datasets/catalog/FAO_GAUL_2015_level1). Precisaremos informar na **FeatureCollection** o Earth Engine Snippet, como `"FAO/GAUL/2015/level1"`.


Para selecionar um Estado em espec√≠fico utilizaremos o comando `Filter` e iremos inserir o nome do Estado (sem acentos). Neste exemplo queremos analisar o Estado do Rio Grande do Sul (RS). Ent√£o iremos inserir o nome `Rio Grande Do Sul`.

```python
# selecionando o contorno do Estado do Rio Grande do Sul
rs = ee.FeatureCollection('FAO/GAUL/2015/level1').filter(ee.Filter.eq('ADM1_NAME', 'Rio Grande Do Sul'))
rs
```

```python
# criando mapa interativo
Map = geemap.Map()

# centra o mapa na regi√£o
Map.centerObject(rs, 6)

# adicionando basemap
Map.add_basemap('SATELLITE')

# contorno de MG
style = {'color': 'yellow', 'fillColor': '00000000'}
Map.addLayer(rs.style(**style), {}, 'MG')

# exibe na tela
Map
```

O GEE tamb√©m possui uma base de dados dos contornos dos munic√≠pios para o mundo todo. Esta base de dados √© uma Feature Collection de [n√≠vel 2](https://developers.google.com/earth-engine/datasets/catalog/FAO_GAUL_2015_level2). Precisaremos informar na **FeatureCollection** o Earth Engine Snippet, como `"FAO/GAUL/2015/level2"`.


Para selecionar um munic√≠pio em espec√≠fico utilizaremos o comando `Filter` e iremos inserir o nome do munic√≠pio (sem acentos). Neste exemplo queremos analisar o munic√≠pio de Porto Alegre (RS). Ent√£o iremos inserir o nome `Porto Alegre`.

```python
# carrega os munic√≠pios a n√≠vel global
municipios_global = ee.FeatureCollection('FAO/GAUL/2015/level2')
municipios_global
```

```python
# seleciona o munic√≠pio de "Porto Alegre (RS)"
porto_alegre = municipios_global.filter(ee.Filter.eq('ADM2_NAME', 'Porto Alegre'))
porto_alegre
```

Iremos utilizar a fun√ß√£o `centerObject` para centralizar nosso mapa no munic√≠pio de Porto Alegre. Como argumento iremos inserir a *FeatureCollection* do munic√≠pio de Porto Alegre e aplicar um zoom de valor 10. Al√©m disso, iremos utilizar a fun√ß√£o `addLayer` para plotar o contorno de Porto Alegre no mapa.

```python
# criando um mapa interativo
Map = geemap.Map()

# centra o mapa na regi√£o
Map.centerObject(porto_alegre, zoom=10)

# adicionando basemap
Map.add_basemap('Esri.WorldImagery')

# contorno de Porto Alegre
style = {'color': 'yellow', 'fillColor': '00000000'}
Map.addLayer(porto_alegre.style(**style), {}, 'itajuba')

# exibe na tela
Map
```

## 2.1 - Carregando e conhecendo os dados

Os dados digitais de eleva√ß√£o de terreno foram gerados pela miss√£o *Shuttle Radar Topography Mission (SRTM)*, atrav√©s de um esfor√ßo de pesquisa internacional que obteve modelos digitais de eleva√ß√£o em uma escala quase global.

O produto SRTM V3 (SRTM Plus) √© fornecido pela NASA JPL com uma resolu√ß√£o de 1 segundo de arco (aproximadamente 30 m). Esse conjunto de dados passou por um processo de preenchimento de vazios usando dados de c√≥digo aberto (ASTER GDEM2, GMTED2010 e NED). Maiores informa√ß√µes acessar [Farr et al. (2007)](https://agupubs.onlinelibrary.wiley.com/doi/10.1029/2005RG000183).

Os dados est√£o dispon√≠veis no [GEE](https://developers.google.com/earth-engine/datasets/catalog/USGS_SRTMGL1_003?hl=pt-br#bands) e s√£o carregados atrav√©s do Snippet `"USGS/SRTMGL1_003"`.

Carregando os dados

```python
srtm = ee.Image('USGS/SRTMGL1_003')
```

Exibindo informa√ß√µes dos dados

```python
# mostrando as informa√ß√µes da imagem
srtm
```

```python
# exibindo as propriedades da imagem
srtm.getInfo()
```

```python
# quantidade de bandas dispon√≠veis
srtm.bandNames().getInfo()
```

```python
# selecionando a eleva√ß√£o
elevacao = srtm.select('elevation')
```

Veja que nos dados do SRTM temos apenas a banda ou vari√°vel `elevation`. Ent√£o iremos utilizar essa vari√°vel.

## 2.2 - Mapas para diferentes regi√µes

###a) Mapa de eleva√ß√£o `global`

Lembrando como cria uma mapa interativo no geemap

```python
Map = geemap.Map()
Map
```

Para plotar o dados de eleva√ß√£o no mapa interativo iremos usar a fun√ß√£o `Map.addLayer()` inserindo como argumento a vari√°vel **srtm** e os par√¢metros de visualiza√ß√£o, como:
 1. **valor m√≠nimo**: 0 metros
 2. **m√°ximo**: 6000 metros
 3. **paleta de cores**: lista com os valores das cores no formato hexadecimal - ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']

Al√©m disso, iremos adicionar uma barra de cores no mapa interativo atrav√©s da fun√ß√£o `add_colorbar()`.

```python
# criando mapa interativo
Map = geemap.Map()

# par√¢metros de visualiza√ß√£o
param_vis = {'min': 0, 'max': 2000, 'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']}

# mapa de eleva√ß√£o
Map.addLayer(elevacao, param_vis, 'Eleva√ß√£o')

# barra de cores
Map.add_colorbar(param_vis, label='Eleva√ß√£o (m)', orientation='vertical')

# exibe na tela
Map
```

###b) Mapa de eleva√ß√£o para um determinado `Pa√≠s`


Vamos aplicar um procedimento semelhante da imagem global, agora apenas para o Brasil. Para isto usaremos a fun√ß√£o `clip()`. Ela usa o contorno da regi√£o e recorta os dados apenas para aquela regi√£o.

```python
# carregando os contornos de todos os pa√≠ses do mundo
paises = ee.FeatureCollection('FAO/GAUL/2015/level0')

# selecionando o contorno do Brasil
brasil = paises.filter(ee.Filter.eq('ADM0_NAME', 'Brazil'))
```

```python
# criando mapa interativo
Map = geemap.Map()

# centraliza o mapa
Map.centerObject(brasil, 4)

# par√¢metros de visualiza√ß√£o
param_vis = {'min': 0, 'max': 2000, 'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']}

# plotando mapa
Map.addLayer(elevacao.mask(elevacao).clip(brasil), param_vis, 'SRTM')

# contorno do Brasil
style = {'color': 'yellow', 'fillColor': '00000000'}
Map.addLayer(brasil.style(**style), {}, 'Brasil')

# barra de cores
Map.add_colorbar(param_vis, label='Eleva√ß√£o (m)', orientation='vertical')

# exibe na tela
Map
```

###e) Mapa de eleva√ß√£o para uma determinada `Geometria`

Imagine que voc√™ gostaria de recortar e plotar o dado de eleva√ß√£o para uma regi√£o delimitada por um ret√¢ngulo. Assim, deve-se informar quais s√£o as latitudes e longitudes que limitam essa geometria.

Esta geometria √© representada (ou delimitada) por uma lista (ou array) que contem quatro n√∫meros. Esses n√∫meros representam as coordenadas geogr√°ficas dos cantos do ret√¢ngulo, na seguinte ordem: `[oeste, sul, leste, norte]`

1. **-55.1**: Longitude do limite oeste (West)
2. **-31.5**: Latitude do limite sul (South)
3. **-50.1**: Longitude do limite leste (East)
4. **-28.4**: Latitude do limite norte (North)

```python
# cria a moldura do mapa
Map = geemap.Map()

# centraliza o mapa
Map.centerObject(retangulo, 7)

# par√¢metros de visualiza√ß√£o
param_vis = {'min': 0, 'max': 2000, 'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']}

# plotando mapa
Map.addLayer(elevacao.mask(elevacao).clip(retangulo), param_vis, 'SRTM')

# barra de cores
Map.add_colorbar(param_vis, label='Eleva√ß√£o (m)', orientation='vertical')

# exibe na tela
Map
```

#**AULA 3):** Mapeamento de Precipita√ß√£o - `AULA REALIZADA`

A precipita√ß√£o √© uma das vari√°veis meteorol√≥gicas mais importantes para deflagra√ß√£o de desastres hidrometeorol√≥gicos. Com o advento dos sat√©lites tornou-se poss√≠vel mapear globalmente a precipita√ß√£o em tempo quase-real. Estas informa√ß√µes s√£o √∫teis para an√°lises climatol√≥gicas anuais e mensais da precipita√ß√£o para uma determinada regi√£o.

Existem diversos produtos de estimativa de precipita√ß√£o por sat√©lite. A escolha depender√° principalmente do seu objetivo, e das caracter√≠sticas de cada produto, como resolu√ß√£o espacial e temporal, e per√≠odo de dados dispon√≠veis. O GEE fornece diversos produtos de estimativa de precipita√ß√£o por sat√©lite, por√©m os principais s√£o:

1.   [Climate Hazards Center InfraRed Precipitation With Station Data (CHIRPS)](https://developers.google.com/earth-engine/datasets/catalog/UCSB-CHG_CHIRPS_DAILY?hl=pt-br): O CHIRPS √© um conjunto de dados de precipita√ß√£o quase global com mais de `30 anos (desde 1981)`. O CHIRPS incorpora imagens de sat√©lite de resolu√ß√£o de `0,05¬∞` com dados de esta√ß√µes in situ para criar s√©ries temporais de precipita√ß√£o em grade para an√°lise de tend√™ncias e monitoramento de seca sazonal. Os dados possuem resolu√ß√£o espacial de `5566 metros` e est√£o dispon√≠veis desde `1981-01-01`. O GEE fornece dados do CHIRPS na frequ√™ncia temporal [di√°ria](https://developers.google.com/earth-engine/datasets/catalog/UCSB-CHG_CHIRPS_DAILY?hl=pt-br) e [pentada](https://developers.google.com/earth-engine/datasets/catalog/UCSB-CHG_CHIRPS_PENTAD?hl=pt-br).

2. [Global Precipitation Measurement (GPM) Release 07 (IMERGE)](https://developers.google.com/earth-engine/datasets/catalog/NASA_GPM_L3_IMERG_V07?hl=pt-br): O IMERGE √© uma miss√£o internacional de sat√©lite que fornece observa√ß√µes de √∫ltima gera√ß√£o de chuva e neve no mundo todo a cada tr√™s horas. O Integrated Multi-satellitE Retrievals for GPM (IMERG) √© o algoritmo unificado que fornece estimativas de precipita√ß√£o combinando dados de todos os instrumentos de micro-ondas passivos na Constela√ß√£o GPM. Os dados possuem resolu√ß√£o espacial de `11132 metros` e est√£o dispon√≠veis desde `2000-06-01`. O GEE fornece dados do IMERGE na frequ√™ncia temporal de [30 min](https://developers.google.com/earth-engine/datasets/catalog/NASA_GPM_L3_IMERG_V07) e [mensal](https://developers.google.com/earth-engine/datasets/catalog/NASA_GPM_L3_IMERG_MONTHLY_V07).

3. [GSMaP Operational: Global Satellite Mapping of Precipitation - V8](https://developers.google.com/earth-engine/datasets/catalog/JAXA_GPM_L3_GSMaP_v8_operational?hl=pt-br): O GSMaP fornece uma taxa de chuva global por hora com uma resolu√ß√£o de 0,1 x 0,1 grau. O GSMaP √© um produto da miss√£o Global Precipitation Measurement (GPM), que fornece observa√ß√µes globais de precipita√ß√£o em intervalos de tr√™s horas. Os valores s√£o estimados usando radi√¥metros passivos de micro-ondas e infravermelho de v√°rias bandas do sat√©lite GPM Core Observatory e com a ajuda de uma constela√ß√£o de outros sat√©lites. Os dados possuem resolu√ß√£o espacial de `11132 metros` e est√£o dispon√≠veis desde `1998-01-01`. O GEE fornece dados do IMERGE na frequ√™ncia temporal de [1 hora](https://developers.google.com/earth-engine/datasets/catalog/JAXA_GPM_L3_GSMaP_v8_operational?hl=pt-br#bands).

Extraindo a s√©rie temporal para o munic√≠pio de Porto Alegre. Precisamos informar os seguintes atributos para a fun√ß√£o `getTimeSeriesByRegion` do eemont:

üü¢ `bands` = 'precipitation'
- Tipo: String ou Lista de Strings
- Descri√ß√£o: Especifica qual(is) banda(s) ser√°(√£o) extra√≠da(s)
- Valor: 'precipitation' = banda de precipita√ß√£o do CHIRPS
- Alternativas: Poderia ser uma lista ['precipitation', 'other_band']

---

üü¢ `reducer` = ee.Reducer.mean()
- Tipo: EE Reducer
- Descri√ß√£o: Define o m√©todo de agrega√ß√£o espacial dos pixels
- Fun√ß√£o: Calcula a m√©dia dos valores de pixels na regi√£o
- Alternativas comuns:
    - ee.Reducer.mean() ‚Üí M√©dia
    - ee.Reducer.median() ‚Üí Mediana
    - ee.Reducer.sum() ‚Üí Soma
    - ee.Reducer.min() ‚Üí Valor m√≠nimo
    - ee.Reducer.max() ‚Üí Valor m√°ximo
    - ee.Reducer.stdDev() ‚Üí Desvio padr√£o

---

üü¢ `scale` = 5000
- Tipo: N√∫mero (metros)
- Descri√ß√£o: Define a escala espacial/resolu√ß√£o para o c√°lculo
- Unidade: Metros (5000 = 5km)
- Import√¢ncia: Determina o tamanho do pixel para agrega√ß√£o
- Efeito: Valores menores = mais detalhado, maiores = mais generalizado
---

```python
# gera serie temporal
serie_temporal = chirps.getTimeSeriesByRegion(geometry = porto_alegre,
                                              bands = 'precipitation',
                                              reducer = ee.Reducer.mean(),
                                              scale = 5000)
```

## Combina√ß√µes de bandas para visualiza√ß√£o de corpos h√≠dricos


A maneira mais simples e r√°pida de identificarmos corpos h√≠dricos nas imagens de sat√©lite √© fazer o uso de combina√ß√£o de bandas espectrais. Para isto utilizaremos os dados do [Landsat-8](https://developers.google.com/earth-engine/datasets/catalog/LANDSAT_LC08_C02_T1_L2) e aplicaremos as seguintes combina√ß√µes:


---
`1. Natural Color (√Ågua Transparente)`:
*   [Red, Green, Blue] ‚Üí [B4, B3, B2]
*   Utilizado para destacar  **Vegeta√ß√£o e √Ågua**
*   **√Ågua limpa**: Azul claro/transparente
*   **√Ågua turva**: Marrom/Verde
---
`2. Falsa Cor para √Ågua (RGB)`:
*   [SWIR1, NIR, Green] ‚Üí [B6, B5, B3]:
*   Utilizado para destacar **√Ågua Limpa/Lagos**
*   **√Ågua aparece**: Preta/Azul escuro
*   **Vegeta√ß√£o**: Verde claro
*   **Solo**: Amarelo
---
`3. Falsa Cor para Inunda√ß√µes`:
*   [SWIR2, SWIR1, Red] ‚Üí [B7, B6, B4]
*   Utilizado para destacar **Rios/√Ågua Turva**
*   **√Ågua**: Azul escuro/Preto
*   **Vegeta√ß√£o inundada**: Vermelho escuro
*   **√Åreas urbanas**: Amarelo claro
---
`4. Falsa Cor Infravermelha`:
*   [NIR, Green, Blue] ‚Üí [B5, B4, B3]
*   Utilizado para  destacar **Vegeta√ß√£o e √Åreas √ömidas**
*   **√Ågua**: Azul/Preto
*   **Vegeta√ß√£o**: Vermelho
---

Carregando os dados e aplicando o fator de escala

```python
# selecionando o contorno do Estado do Rio Grande do Sul
rs = ee.FeatureCollection('FAO/GAUL/2015/level1').filter(ee.Filter.eq('ADM1_NAME', 'Rio Grande Do Sul'))

# aplica fator de escala
def fator_escala(img):

    # Aplica fatores de escala
    opticas = img.select('SR_B.*').multiply(2.75e-05).add(-0.2)
    termais = img.select('ST_B.*').multiply(0.00341802).add(149.0)

    # Retorna imagem com bandas escaladas
    return (img.addBands(opticas, None, True)
               .addBands(termais, None, True)
               .copyProperties(img, ['system:time_start'])
               .set('date', img.date().format('YYYY-MM-dd')))

# carrega os dados
landsat8 = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
             .filterBounds(rs) \
             .filterDate('2024-01-01','2025-01-01') \
             .filter(ee.Filter.lt('CLOUD_COVER', 5)) \
             .map(fator_escala)

# mostra os dados
landsat8
```

Plotando mapa

```python
# par√¢metros de visualiza√ß√£o
param_vis_1 = {'bands': ['SR_B4', 'SR_B3', 'SR_B2'], 'min': 0.03, 'max': 0.19}
param_vis_2 = {'bands': ['SR_B6', 'SR_B5', 'SR_B3'], 'min': 0.03, 'max': 0.19}
param_vis_3 = {'bands': ['SR_B7', 'SR_B6', 'SR_B4'], 'min': 0.03, 'max': 0.19}
param_vis_4 = {'bands': ['SR_B5', 'SR_B4', 'SR_B3'], 'min': 0.03, 'max': 0.19}

# monta painel com 4 mapas
geemap.linked_maps(rows = 2,
                   cols = 2,
                   height = "400px",
                   center = [-29.94, -51.32],
                   zoom = 11,
                   ee_objects = [landsat8.mean().clip(rs), landsat8.mean().clip(rs), landsat8.mean().clip(rs), landsat8.mean().clip(rs)],
                   vis_params = [param_vis_1, param_vis_2, param_vis_3, param_vis_4],
                   labels = ['Natural Color (√Ågua Transparente)', 'Falsa Cor para √Ågua (RGB)', 'Falsa Cor para Inunda√ß√µes', 'Falsa Cor Infravermelha'],
                   label_position = "topright")
```

## Normalized Difference Water Index (NDWI)

O Normalized Difference Water Index (NDWI) √© um √≠ndice usado para identificar e monitorar corpos d'√°gua. O NDWI foi desenvolvido por McFeeters em 1996 ([The use of NDWI in the delineation of open water features. Int. J. Remote Sensing, 17(7), 1425‚Äì1432](https://www.tandfonline.com/doi/abs/10.1080/01431169608948714)) e emprega a banda do verde (0,5 ¬µm, GREEN) e infravermelho pr√≥ximo (0,8 ¬µm, NIR) para real√ßar as caracter√≠sticas de corpos h√≠dricos, como: rios, lagos, represas e √°reas alagadas. A seguinte equa√ß√£o √© utilizada para o c√°lculo do NDWI:

---

$$
NDWI = \frac{GREEN - NIR}{GREEN + NIR}
$$

---

Em que:
 - `GREEN:` reflet√¢ncia no canal de 0,5 ¬µm  
 - `NIR:` reflet√¢ncia no canal de 0,8 ¬µm


---


Os valores de NDWI variam entre -1 e +1:
 - üåä √Ågua ‚Üí valores altos/positivos, pois reflete intensamente no verde e absorve no NIR.
 - üå± Vegeta√ß√£o e solo ‚Üí valores baixos ou negativos, j√° que refletem mais no NIR que no verde.



Em resumo:
*   `-1,0 √†  -0,3:` seca, superf√≠ces n√£o-aquosas
*   `-0,3 √†   0,0:` seca moderada, superf√≠ces n√£o-aquosas
*   ` 0,0 √†  +0,2:` Alagamento, umidade
*   `+0,2 √†  +1,0:` √Ågua
    
O uso do NDWI √© essencial em diversas √°reas:
- Delimita√ß√£o de lagos, rios e reservat√≥rios
- Monitoramento de cheias e inunda√ß√µes
- Acompanhamento da varia√ß√£o h√≠drica ao longo do tempo
- Apoio √† gest√£o ambiental e ao planejamento agr√≠cola

Vamos calcular o NDWI com dados do sat√©lite [Landsat-8](https://developers.google.com/earth-engine/datasets/catalog/LANDSAT_LC08_C02_T1_L2?hl=pt-br). Para o c√°lculo do NDWI iremos usar as bandas `SR_B3` (0,533 a 0,590 Œºm, reflect√¢ncia da superf√≠cie da banda 3-verde) e `SR_B5` (0,851 a 0,879 Œºm, reflect√¢ncia da superf√≠cie da banda 5-infravermelho pr√≥ximo).

```python
# selecionando o contorno do Estado do Rio Grande do Sul
rs = ee.FeatureCollection('FAO/GAUL/2015/level1').filter(ee.Filter.eq('ADM1_NAME', 'Rio Grande Do Sul'))

# carrega os dados
landsat8 = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
             .filterBounds(rs) \
             .filterDate('2024-01-01','2025-01-01') \
             .filter(ee.Filter.lt('CLOUD_COVER', 5))

# mostra os dados
landsat8
```

Fun√ß√£o que calcula o NDWI

```python
def ndwi(img):

    # fator de escala das bandas √≥ticas
    bandas_oticas = img.select('SR_B.*').multiply(2.75e-05).add(-0.2)

    # fator de escala de bandas termais
    bandas_termais = img.select('ST_B.*').multiply(0.00341802).add(149.0)

    # NDWI
    ndwi_image = img.normalizedDifference(['SR_B3', 'SR_B5']).rename('ndwi')

    # aplicando e adicionando as bandas com fator de escala
    return img.addBands(bandas_oticas, None, True) \
              .addBands(bandas_termais, None, True) \
              .addBands(ndwi_image) \
              .copyProperties(img, ['system:time_start']) \
              .set('date', img.date().format('YYYY-MM-dd'))
```

Aplica a fun√ß√£o de NDWI

```python
# aplica a fun√ß√£o de NDWI em cada imagem
imagens_ndwi = landsat8.map(ndwi)

# mostra as imagens
imagens_ndwi
```

Cria as m√°scaras para identificar valores `m√°ximos` e `m√©dios` de NDWI

```python
mask_maximo = imagens_ndwi.select('ndwi').max().gt(0).selfMask()
mask_medio = imagens_ndwi.select('ndwi').mean().gt(0).selfMask()
```

#**AULA 6):** Mapeamento da Turbidez da √Ågua - `AULA REALIZADA`

Iremos determinar a turbidez da √°gua utilizando o `Normalized Difference Turbidity Index (NDTI)`. O NDTI √© o √≠ndice de turbidez da √°gua que mede a concentra√ß√£o de sedimentos em suspens√£o. A turbidez indica a "opacidade" da √°gua causada por part√≠culas em suspens√£o.



---


**Equa√ß√£o:**
- O NDTI √© calculado combinando a reflect√¢ncia do comprimento de onda do vermelho (0.65 Œºm)e do verde (0.55 Œºm), atrav√©s da seguintes equa√ß√£o:


- `NDTI = (Red - Green) / (Red + Green)`

Onde:
- Red = Banda do vermelho (‚âà 0.65 Œºm)
- Green = Banda do verde (‚âà 0.55 Œºm)

---


**Interpreta√ß√£o dos Valores:**
- `-1.0 a -0.1:` √Ågua limpa	-> **Baixa turbidez**
- `-0.1 a  0.1:` √Ågua moderadamente clara	-> **Turbidez moderada**
- ` 0.1 a  0.3:` √Ågua turbulenta -> **Alta turbidez**
- ` 0.3 a  1.0:` √Ågua muito turva	-> **Turbidez muito alta**


---
**Princ√≠pio F√≠sico:**
- `√Ågua limpa:` Absorve mais luz vermelha, reflete mais verde
- `√Ågua turva:` Part√≠culas em suspens√£o espalham a luz, aumentando reflex√£o no vermelho


---
**Aplica√ß√µes:**
- Controle de eros√£o em bacias hidrogr√°ficas
- Monitoramento de descarga de sedimentos
- Impacto de atividades agr√≠colas e minera√ß√£o

---

**Vantagens do NDTI:**
- Simples de calcular
- Eficaz para detectar sedimentos em suspens√£o
- Compat√≠vel com m√∫ltiplos sat√©lites
- Bom indicador de qualidade da √°gua

**O NDTI √© uma ferramenta poderosa para monitoramento da turbidez da √°gua usando sensoriamento remoto**.

### Carregando os dados, aplicando o mascaramento de nuvens e calculando o NDTI

Iremos utilizar o seguintes dados:

1. [Harmonized Sentinel-2 MSI: MultiSpectral Instrument, Level-2A (SR)](https://developers.google.com/earth-engine/datasets/catalog/COPERNICUS_S2_SR_HARMONIZED?hl=pt-br) do Sentinel-2.

2. [Sentinel-2: Cloud Probability](https://developers.google.com/earth-engine/datasets/catalog/COPERNICUS_S2_CLOUD_PROBABILITY?hl=pt-br)

```python
# carregando os dados
sen2 = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED").linkCollection(ee.ImageCollection("COPERNICUS/S2_CLOUD_PROBABILITY"), 'probability') \
         .filterBounds(regiao) \
         .filterDate('2024', '2025')
sen2
```

Veja que a banda de probabilidade de nuvens foi adicionada ao dataset

image.png

A fun√ß√£o abaixo calcula o NDTI com os dados do sentinel 2, usando as bandas `B4 (vermelho)` e `B3 (verde)`.


NDTI = red - green / red + green

```python
# criando uma fun√ß√£o que calcula o NDTI
def turbidez(img):

    # seleciona imagens com menos de 20% de nuvens
    cloud = img.select('probability')
    cloud_free = cloud.lt(20)

    # aplica o fator de escala do das bandas de reflect√¢ncia
    sr = img.select('B.*').multiply(0.0001)

    # calcula o NDWI e seleciona valores acima de 0.1
    ndwi = sr.normalizedDifference(['B3', 'B8']).rename('ndwi')
    water_body = ndwi.gt(0.1)

    # calcula o NDTI
    ndti = sr.normalizedDifference(['B4', 'B3']).rename('ndti')

    return ndti.updateMask(cloud_free).updateMask(water_body).copyProperties(img, ['system:time_start'])
```

```python
# carregando os dados
sen2 = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED").linkCollection(ee.ImageCollection("COPERNICUS/S2_CLOUD_PROBABILITY"), 'probability') \
         .filterBounds(regiao) \
         .filterDate('2024', '2025') \
         .map(turbidez)
sen2
```

### Transformando as imagens para o formato do dataset do xarray

O atributo escala √© a resolu√ß√£o espacial em graus. As bandas `B4 (vermelho)` e `B8 (verde)` do Sentinel-2 tem 10 m de resolu√ß√£o espacial. Em km corresponde a 0.01 km. Para transformar de km para graus dividimos por 100, ou seja, 0.01/100 = 0.0001 graus.

- `10 m = 0.0001 graus`


```python
# usando 100 m de resolu√ß√£o espacial
ds = xr.open_dataset(sen2,
                     engine = 'ee',
                     crs = 'EPSG:4326',
                     scale = 0.001,
                     geometry = regiao)
```

## Imagens `MNDWI do Sentinel-2`

O NDWI emprega as bandas do verde (GREEN) e infravermelho pr√≥ximo (NIR) para detectar corpos h√≠dricos. O NDWI identifica bem corpos h√≠dricos para √°guas claras, mas em ambientes com turbidez elevada ou interfer√™ncia urbana, ele tende a subestimar a presen√ßa de √°gua.

Ap√≥s uma inunda√ß√£o √© comum que a exist√™ncia de alta concentra√ß√£o de sedimentos supensos na √°gua com elevada turbidez, altere a assinatura espectral da √°gua, principalmente na banda do infravermelho pr√≥ximo. Este comportamento tende a reduzir a difer√™n√ßa de reflect√¢ncia entre √°gua e solo, dificultando assim a identifica√ß√£o pelo NDWI das √°reas realmente inundadas.

Uma alternativa seria empregarmos o Modified Normalized Difference Water Index (MNDWI). Proposto por Hanqiu Xu (2006) ([Modification of NDWI to enhance open water features in remotely sensed imagery. Int. J. Remote Sensing, 27(14), 3025‚Äì3033](https://www.tandfonline.com/doi/full/10.1080/01431160600589179)), o MNDWI substitui a banda do infravermelho pr√≥ximo (NIR) pela banda do infravermelho de ondas curtas (SWIR). Essa modifica√ß√£o melhora o contraste espectral entre √°gua e demais alvos, permitindo uma discrimina√ß√£o mais eficiente de superf√≠cies inundadas, mesmo em √°reas urbanizadas ou com alta carga de sedimentos. √â um importante √≠ndice para suprimir o ru√≠do de √°reas constru√≠das que os √≠ndices de √°gua podem confundir.

A seguinte equa√ß√£o √© utilizada para o c√°lculo do MNDWI:

---

$$
MNDWI = \frac{GREEN - SWIR}{GREEN + SWIR}
$$

---

Em que:
 - `GREEN:` reflet√¢ncia no canal de 0,5 ¬µm  
 - `SWIR:` reflet√¢ncia no canal de 1,6 ¬µm


---


`Interpreta√ß√£o dos valores:`
*   Valores positivos geralmente indicam a presen√ßa de √°gua.
*   Valores negativos geralmente indicam √°reas urbanas, vegeta√ß√£o ou solo.


`Vantagens:`
- O MNDWI √© mais eficaz na detec√ß√£o de corpos d'√°gua em √°reas urbanas do que o NDWI (√çndice da Diferen√ßa Normalizada da √Ågua), pois o uso da banda SWIR ajuda a suprimir o reflexo de telhados e asfalto.

`Aplica√ß√µes:`
- Monitoramento de inunda√ß√µes
- Gerenciamento de recursos h√≠dricos
- Mapeamento de √°reas alagadas
- An√°lise de √°reas √∫midas e ecossistemas costeiros
- Monitoramento da disponibilidade h√≠drica na agricultura

Para o sat√©lite Sentinel-2 usaremos a banda `B3` (560nm (S2A) / 559nm (S2B), verde) e `B11` (1.613,7 nm (S2A) / 1.610,4 nm (S2B), SWIR 1).

```python
#==============================================================================#
#                             CARREGA OS DADOS
#==============================================================================#
# regi√£o de interesse
regiao = ee.Geometry.Rectangle([-51.65, -30.09, -50.90, -29.81])

# carrega os dados para o per√≠odo todo
sen2 = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
         .filterBounds(regiao) \
         .filterDate('2024-04-01','2024-06-01') \
         .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 5)) \
         .map(lambda img: img.multiply(0.0001).copyProperties(img, img.propertyNames()))

# imagem ANTES
imagem_antes = sen2.filterDate('2024-04-01', '2024-04-21').mean()

# imagem DEPOIS
imagem_depois = sen2.filterDate('2024-05-06', '2024-05-30').mean()

#==============================================================================#
#                           C√ÅLCULO DE MNDWI
#==============================================================================#
# "NDWI" usa a banda do VERDE e INFRAVERMELHO. Real√ßa os corpos h√≠dricos.
# Locais com √°gua fica em tons mais escuros (=1), e o que n√£o √© √°gua em tons mais claro (0).

# Em adi√ß√£o o "MNDWI" real√ßa os corpos h√≠dricos e retira alguns ru√≠dos
# No "MNDWI" todo pixel com valor maior que 0 tem √°gua. Os valores de MNDWI entre 0 a 1, indicam √°gua. Assim
# a grada√ß√£o entre 0 e 1 indicam grau de profundidade da l√¢mina de √°gua. Quanto mais pr√≥ximo de 1 maior a profundidade da
# l√¢mina de √°gua. As nuvens s√£o mascaradas, pois pegaremos apenas MNDWI positivos, e nuvens tem valores negativos.

#---------------------------------------#
#            Antes da chuva
#          (01-21/abril/2024)
#---------------------------------------#
mndwi_antes = imagem_antes.normalizedDifference(['B3', 'B11']).rename('mndwi')
mask_antes = mndwi_antes.gte(0)
mndwi_antes_mask = mndwi_antes.updateMask(mask_antes)

#---------------------------------------#
#           Depois da chuva
#          (06-30/maio/2024)
#---------------------------------------#
mndwi_depois = imagem_depois.normalizedDifference(['B3', 'B11']).rename('mndwi')
mask_depois = mndwi_depois.gte(0)
mndwi_depois_mask = mndwi_depois.updateMask(mask_depois)
```

Calcula a √°rea de inunda√ß√£o

```python
#==============================================================================#
#                          C√ÅLCULO DA √ÅREA (KM¬≤)
#==============================================================================#
# imagem com a mascara para √°gua
agua_antes_mask = mndwi_antes.gte(0).selfMask().rename('antes').set('date', '2024-04-21')
agua_depois_mask = mndwi_depois.gte(0).selfMask().rename('depois').set('date', '2024-05-06')

# junta as imagens com √°gua ANTES e DEPOIS
mask_area = agua_antes_mask.addBands(agua_depois_mask)

# c√°lcula a √°rea de cada pixel em km¬≤:
# 1. Multiplica pela √°rea do pixel em m¬≤
# 2. Divide por 1.000.000 para converter para km¬≤
imagem_area_km2 = mask_area.multiply(ee.Image.pixelArea()).divide(1000000)

# acumula todos os pixels que est√£o em km¬≤
soma_area_km2 = imagem_area_km2.reduceRegion(reducer=ee.Reducer.sum(),
                                             geometry=regiao,
                                             scale=10,
                                             maxPixels=1e13,
                                             bestEffort=True)

# obt√©m os valores em km¬≤
area_antes_km2 = soma_area_km2.get('antes').getInfo()
area_depois_km2 = soma_area_km2.get('depois').getInfo()

# ==============================================================================#
#                         IMPRESS√ÉO DOS RESULTADOS (KM¬≤)
# ==============================================================================#

print("=" * 70)
print("üìä AN√ÅLISE DE √ÅREA DE √ÅGUA - ANTES vs DEPOIS")
print("=" * 70)

print(f"\nüìÖ DATAS:")
print(f"   ‚Ä¢ Antes do evento: 2024-04-21")
print(f"   ‚Ä¢ Depois do evento: 2024-05-06")

print(f"\nüó∫Ô∏è √ÅREA EM QUIL√îMETROS QUADRADOS (km¬≤):")
print(f"   ‚úÖ Antes:  {area_antes_km2:,.2f} km¬≤")
print(f"   ‚úÖ Depois: {area_depois_km2:,.2f} km¬≤")

print(f"\nüìà VARIA√á√ÉO DA √ÅREA:")
if area_antes_km2 and area_antes_km2 > 0:  # Verifica se n√£o √© None e > 0
    variacao_km2 = area_depois_km2 - area_antes_km2
    variacao_percentual = (variacao_km2 / area_antes_km2) * 100

    print(f"   üìä Varia√ß√£o absoluta: {variacao_km2:+,.2f} km¬≤")
    print(f"   üìà Varia√ß√£o percentual: {variacao_percentual:+.2f}%")

    if variacao_km2 > 0:
        print(f"   üíß Status: AUMENTO de {variacao_km2:,.2f} km¬≤ na √°rea de √°gua")
    elif variacao_km2 < 0:
        print(f"   üèúÔ∏è Status: REDU√á√ÉO de {abs(variacao_km2):,.2f} km¬≤ na √°rea de √°gua")
    else:
        print(f"   ‚öñÔ∏è Status: √Årea mantida SEM ALTERA√á√ïES")
else:
    print(f"   ‚ö†Ô∏è N√£o foi poss√≠vel calcular a varia√ß√£o (√°rea inicial: {area_antes_km2} km¬≤)")

print("\n" + "=" * 70)
```

O sat√©lite [Sentinel-1](https://space.oscar.wmo.int/satelliteprogrammes/view/sentinel_1) possui a bordo um `Radar de Abertura Sint√©tica` (SAR) operando em Banda-C (5,4 GHz) com dupla polariza√ß√£o, permitindo assim avaliar inunda√ß√µes mesmo na presen√ßa de nuvens. Vejam exemplos [aqui](https://www.un-spider.org/advisory-support/recommended-practices/recommended-practice-flood-mapping/step-by-step). Os dados dispon√≠veis no [GEE](https://developers.google.com/earth-engine/datasets/catalog/COPERNICUS_S1_GRD) possuem frequ√™ncia temporal de 6 dias, e cada cena possui uma das 3 resolu√ß√µes (10, 25 ou 40 metros), 4 combina√ß√µes de bandas (correspondentes √† polariza√ß√£o da cena) e 3 modos de instrumento.

Cada cena contem 1 ou 2 das 4 bandas de polariza√ß√£o poss√≠veis, dependendo das configura√ß√µes de polariza√ß√£o do instrumento. As combina√ß√µes poss√≠veis s√£o: banda √∫nica VV, banda √∫nica HH, banda dupla VV+VH e banda dupla HH+HV.

1. `VV:` co-polariza√ß√£o √∫nica, transmiss√£o vertical/recep√ß√£o vertical
2. `HH:` co-polariza√ß√£o √∫nica, transmiss√£o horizontal/recep√ß√£o horizontal
3. `VV + VH:` polariza√ß√£o cruzada de banda dupla, transmiss√£o vertical/recep√ß√£o horizontal
4. `HH + HV:` polariza√ß√£o cruzada de banda dupla, transmiss√£o horizontal/recep√ß√£o vertical

Tipicamente espalhamento de superf√≠cie, como solo nu e √°gua, s√£o mais sensiveis para espalhamento com polariza√ß√£o VV. Portanto, polariza√ß√£o VV √© frequentemente utilizada para detectar corpos d'√°gua.

Sentinel-1 opera em 4 modos de aquisi√ß√£o:
- Stripmap (SM)
- Interferometric Wide swath (IW)
- Extra-Wide swath (EW)
- Wave mode (WV)

O modo `Interferometric Wide swath (IW)` permite combinar um largo swath (250 km) com uma moderada resolu√ß√£o geom√©trica (5 √† 20 m). O modo Interferometric Wide swath (IW) √© o modo de aquisi√ß√£o padr√£o sobre a terra. N√≥s usaremo o modo IW do Sentinel-1 para extrair a extens√£o da inunda√ß√£o.

```python
# regi√£o de interesse
regiao = ee.Geometry.Rectangle([-51.65, -30.09, -50.90, -29.81])

# carrega os dados para abril e maio de 2024
sen1 = ee.ImageCollection('COPERNICUS/S1_GRD') \
         .filterBounds(regiao) \
         .filterDate('2024-04-01','2024-06-01')
```

Na regi√£o tropical, as imagens do Sentinel 1 possuem duas bandas:
- VV: polariza√ß√£o singular (vertical-vertical);
- VH: dupla polariza√ß√£o (vertical-horizontal).

J√° nas regi√µes polares, outras duas bandas s√£o importantes:
- HH: polariza√ß√£o singular (horizontal-horizontal);
- HV: dupla polariza√ß√£o (horizontal-vertical).

Um RGB comum dos sensores √≥ticos √© imposs√≠vel de ser feito em imagens de Radares de Abertura Sint√©tica. Mesmo assim, uma pr√°tica comum √© criar um RGB Sint√©tico combinando as bandas VV, VH e a raz√£o VV/VH entre elas.

```python
VV = sen1.select('VV').mean().clip(regiao)
VH = sen1.select('VH').mean().clip(regiao)
RGB = VV.addBands(VV).addBands(VV.divide(VH))
```

Plota mapa interativo: `layers`

Carregando os dados

- O `speckle`, que aparece nas imagens de radar de abertura sint√©tica (SAR) como ru√≠do granular, √© causado pela interfer√™ncia de ondas refletidas por muitos espalhadores elementares. O speckle nas imagens SAR complica a interpreta√ß√£o das imagens, reduzindo a efic√°cia da segmenta√ß√£o e classifica√ß√£o das imagens ([Lee et al., 1994](https://www.tandfonline.com/doi/abs/10.1080/02757259409532206)). Portanto, a filtragem de speckle √© frequentemente aplicada a imagens SAR para reduzir o ru√≠do de speckle. Neste exemplo, aplicamos um filtro morfol√≥gico de speckle √†s imagens SAR do Sentinel-1. O filtro morfol√≥gico de speckle √© um filtro n√£o linear que usa o valor mediano de um pixel e seus pixels vizinhos para substituir o valor do pixel. O tamanho do kernel √© definido como 100 metros.



```python
# per√≠odo ANTES da inunda√ß√£o
antes_data_inicial, antes_data_final = '2023-10-01', '2024-04-01'

# per√≠odo DURANTE/DEPOIS da inunda√ß√£o
depois_data_inicial, depois_data_final = '2024-04-22', '2024-05-15'

# selecionando o contorno do Estado do Rio Grande do Sul
regiao = ee.FeatureCollection('FAO/GAUL/2015/level1').filter(ee.Filter.eq('ADM1_NAME', 'Rio Grande Do Sul'))

# dados ANTES
sen1_antes = ee.ImageCollection("COPERNICUS/S1_GRD") \
               .filter(ee.Filter.listContains("transmitterReceiverPolarisation", "VV")) \
               .filter(ee.Filter.eq("instrumentMode", "IW")) \
               .filterDate(antes_data_inicial, antes_data_final) \
               .filterBounds(regiao) \
               .select("VV")

# dados DEPOIS
sen1_depois = ee.ImageCollection("COPERNICUS/S1_GRD") \
                .filter(ee.Filter.listContains("transmitterReceiverPolarisation", "VV")) \
                .filter(ee.Filter.eq("instrumentMode", "IW")) \
                .filterDate(depois_data_inicial, depois_data_final) \
                .filterBounds(regiao) \
                .select("VV")

# aplica filtro de speckle
sen1_antes = sen1_antes.map(lambda img: img.focal_median(100, "circle", "meters"))
sen1_depois = sen1_depois.map(lambda img: img.focal_median(100, "circle", "meters"))

# mostra a quantidade de imagens carregadas.
print(f"Quantidade de imagens antes da inunda√ß√£o: {sen1_antes.size().getInfo()}")
print(f"Quantidade de imagens depois da inunda√ß√£o: {sen1_depois.size().getInfo()}")

# cria uma imagem representativa do per√≠odo "antes" usando o percentil 20 e recorta para a regi√£o de estudo
# aplica o percentil 20 (20th percentile) em uma cole√ß√£o temporal, reduzindo m√∫ltiplas imagens temporais em uma √∫nica imagem.
# Exemplo: Se voc√™ tem 100 imagens de radar para o per√≠odo "antes". Para cada pixel, pega o valor do 20¬∫ percentil temporal
# Resultado: 1 imagem com o valor do percentil 20 para cada pixel
# vantagens: reduz efeito de speckle (ru√≠do do radar), minimiza varia√ß√µes temporais, representa condi√ß√µes "baixas" (√∫til para linha de base) e remove picos de retroespalhamento
sen1_antes = sen1_antes.reduce(ee.Reducer.percentile([20])).clipToCollection(regiao)
sen1_depois = sen1_depois.reduce(ee.Reducer.percentile([20])).clipToCollection(regiao)
```

- A √°gua geralmente aparece escura nas imagens SAR porque as ondas eletromagn√©ticas do radar s√£o refletidas de maneira diferente por diferentes superf√≠cies. A √°gua √© uma superf√≠cie lisa e plana que n√£o reflete muito bem as ondas de radar, por isso aparece escura nas imagens SAR. O limiar das imagens SAR √© uma das abordagens mais utilizadas para delinear a extens√£o da √°gua devido √† sua efic√°cia e efici√™ncia ([Liang e Liu, 2020](https://www.sciencedirect.com/science/article/pii/S0924271619302540?via%3Dihub)). Os m√©todos aplicando limiares podem ser geralmente divididos em duas categorias: global e local. Os m√©todos de limiar global utilizam um √∫nico valor de limiar para segmentar toda a imagem. Os m√©todos de limiar local utilizam um valor de limiar diferente para cada pixel. Neste exemplo, utilizamos um m√©todo de limiar global para extrair a extens√£o da √°gua. O valor do limiar √© definido para `-18 dB`.

```python
ano_inicial = ee.Date('2010')
ano_final = ee.Date('2025')
dif_tempo = ano_final.difference(ano_inicial, 'month').round()
lista_tempo = ee.List.sequence(0, ee.Number(dif_tempo).subtract(1), 1).map(lambda x: ano_inicial.advance(x, 'month'))
lista_tempo
```

```python
# carrega os dados do CHIRPS
chirps_2020_a_2024 = ee.ImageCollection('UCSB-CHG/CHIRPS/PENTAD') \
                       .filterDate(ano_inicial, ano_final)\
                       .select('precipitation') \
                       .map(lambda img: img.clip(mg).copyProperties(img, img.propertyNames()))

# mostra os dados
chirps_2020_a_2024
```

```python
def mensal(data):
    data_inicial = ee.Date(data)
    data_final = data_inicial.advance(1, 'month')
    img = chirps_2020_a_2024.filterDate(data_inicial, data_final).sum()
    return img.set('system:time_start', data_inicial.millis())
```

```python
chirps_mensal = ee.ImageCollection(lista_tempo.map(mensal))
chirps_mensal
```

```python
chuva_media = chirps_mensal.mean()
chuva_desvio = chirps_mensal.reduce(ee.Reducer.stdDev())
```
